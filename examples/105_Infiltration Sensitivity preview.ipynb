# Example Functions for Mannings 'n' Sensitivity Analysis

The notebook provides four comprehensive functions for analyzing the sensitivity of HEC-RAS hydraulic models to changes in Manning's roughness coefficient:

1. **`manning_sensitivity_bulk()`**
   - Performs uniform adjustments to all Manning's n values throughout the model
   - Modifies both base overrides and regional overrides simultaneously
   - Useful for understanding overall model response to roughness changes

2. **`manning_sensitivity_baseoverride()`**
   - Adjusts only the base Manning's n values while keeping regional overrides constant
   - Helps isolate the effects of background/default roughness changes
   - Useful for understanding the impact of modifications to the base roughness map

3. **`manning_sensitivity_region()`**
   - Adjusts only regional Manning's n overrides while keeping base values unchanged
   - Can target specific region tables or modify all region tables
   - Useful for understanding the impact of channel roughness modifications

4. **`manning_sensitivity_base_landuse()` and `manning_sensitivity_region_landuse()`**
   - Allow for targeted adjustments to specific land use types in either base or regional overrides
   - Enable detailed sensitivity testing of individual land cover classifications
   - Useful for calibration efforts focused on specific terrain types

Each function follows a similar workflow:
- Creates a series of model runs with incremental adjustments to Manning's n values
- Extracts results at a user-specified point of interest
- Produces visualizations comparing water surface elevations across scenarios
- Calculates sensitivity metrics (change in water surface per unit change in Manning's n)
- Exports results to CSV files and generates summary plots

The functions include features such as:
- Support for both parallel and sequential computation
- Automatic backup of geometry files
- Detailed logging and error handling
- Customizable adjustment increments and ranges
- Visualization of results with color-coded plots
- Quantification of sensitivity relationships

These tools help modelers quantify how Manning's roughness affects model results, support calibration efforts, and document the sensitivity of hydraulic model outputs to roughness parameter changes.

## Bulk Mannings Sensitivity

### manning_sensitivity_bulk

```python
def manning_sensitivity_bulk(
    template_plan,
    point_of_interest,
    interval=0.005,
    num_intervals=4,
    run_parallel=True,
    max_workers=2,
    num_cores=2,
    results_folder="mannings_sensitivity_results"
):
    """
    Perform a sensitivity analysis by uniformly adjusting all Manning's roughness values.
    
    Parameters:
    -----------
    template_plan : str
        The plan number to use as a template (e.g., "01")
    point_of_interest : tuple or Point
        Coordinates (x, y) or Point object for finding the nearest mesh cell
    interval : float, optional
        The increment for adjusting Manning's roughness. Default is 0.005.
    num_intervals : int, optional
        Number of intervals to test in each direction (+/-). Default is 4.
    run_parallel : bool, optional
        Whether to run the plans in parallel. Default is True.
    max_workers : int, optional
        Maximum number of parallel workers. Default is 2.
    num_cores : int, optional
        Number of cores per worker. Default is 2.
    results_folder : str, optional
        Name of the folder to save results. Default is "mannings_sensitivity_results".
    
    Returns:
    --------
    dict
        Dictionary containing the sensitivity analysis results
    """
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path
    from shapely.geometry import Point
    import os
    import time
    
    # Convert point_of_interest to Point if it's not already
    if not isinstance(point_of_interest, Point):
        point_of_interest = Point(point_of_interest[0], point_of_interest[1])
    
    # Create results directory
    results_dir = Path(ras.project_folder) / results_folder
    results_dir.mkdir(exist_ok=True)
    
    # Get the geometry number associated with the template plan
    template_geom = ras.plan_df.loc[ras.plan_df['plan_number'] == template_plan, 'geometry_number'].values[0]
    print(f"Template geometry number: {template_geom}")
    
    # Get the geometry file path
    geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == template_geom, 'full_path'].values[0]
    
    # Get the original Manning's values
    original_baseoverrides = RasGeo.get_mannings_baseoverrides(geom_path)
    original_regionoverrides = RasGeo.get_mannings_regionoverrides(geom_path)
    
    # Store the base plan details for reference
    base_plan_details = {
        'plan_number': template_plan,
        'geom_number': template_geom,
        'adjustment': 0.0,
        'shortid': f"Original"
    }
    
    # Create a list to store all the new plan details
    sensitivity_plans = [base_plan_details]
    
    # Generate plans with adjusted Manning's values
    adjustments = []
    
    # Add positive adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(i * interval)
    
    # Add negative adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(-i * interval)
    
    # Sort adjustments
    adjustments.sort()
    
    # Create plans for each adjustment
    for adjustment in adjustments:
        # Format the adjustment string (e.g., P005 for +0.005, N005 for -0.005)
        if adjustment > 0:
            adj_str = f"P{int(adjustment*1000):03d}"
        else:
            adj_str = f"N{int(abs(adjustment)*1000):03d}"
        
        # Create a new plan with adjusted values
        print(f"\nCreating plan with {adjustment:+.3f} Manning's adjustment (MB_{adj_str})...")
        
        # Clone the template plan and geometry
        new_plan_number = RasPlan.clone_plan(template_plan, new_plan_shortid=f"MB_{adj_str}")
        new_geom_number = RasPlan.clone_geom(template_geom)
        
        # Set the new plan to use the new geometry
        RasPlan.set_geom(new_plan_number, new_geom_number)
        
        # Get the new geometry file path
        new_geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == new_geom_number, 'full_path'].values[0]
        
        # Create adjusted Manning's values
        modified_baseoverrides = original_baseoverrides.copy()
        modified_baseoverrides["Base Manning's n Value"] = modified_baseoverrides["Base Manning's n Value"] + adjustment
        
        # Apply base Manning's adjustments
        RasGeo.set_mannings_baseoverrides(new_geom_path, modified_baseoverrides)
        
        # Only apply region override adjustments if they exist
        if not original_regionoverrides.empty:
            modified_regionoverrides = original_regionoverrides.copy()
            modified_regionoverrides['MainChannel'] = modified_regionoverrides['MainChannel'] + adjustment
            RasGeo.set_mannings_regionoverrides(new_geom_path, modified_regionoverrides)
        
        # Store the plan details
        sensitivity_plans.append({
            'plan_number': new_plan_number,
            'geom_number': new_geom_number,
            'adjustment': adjustment,
            'shortid': f"MB_{adj_str}"
        })
    
    # Get all plan numbers for execution (excluding the template which is already computed)
    plan_numbers = [plan_details['plan_number'] for plan_details in sensitivity_plans 
                   if plan_details['plan_number'] != template_plan]
    
    # Run the plans (either in parallel or sequentially)
    if run_parallel:
        print(f"\nRunning {len(plan_numbers)} plans in parallel...")
        results = RasCmdr.compute_parallel(
            plan_number=plan_numbers,
            max_workers=max_workers,
            num_cores=num_cores,
            clear_geompre=True
        )
    else:
        print(f"\nRunning {len(plan_numbers)} plans sequentially...")
        results = {}
        for plan_number in plan_numbers:
            print(f"  Running plan {plan_number}...")
            result = RasCmdr.compute_plan(plan_number, clear_geompre=True)
            results[plan_number] = result
    
    print("\nExecution results:")
    for plan, success in results.items():
        print(f"  Plan {plan}: {'Successful' if success else 'Failed'}")
    
    # Check if we need to get the geometry HDF path
    geom_hdf_path = None
    # Use template geometry for mesh cell identification
    geom_number = template_geom
    geom_hdf_path = ras.geom_df.loc[ras.geom_df['geom_number'] == geom_number, 'hdf_path'].values[0]
    
    # Find the nearest mesh cell to the point of interest
    mesh_cells_gdf = HdfMesh.get_mesh_cell_points(geom_hdf_path)
    
    # Calculate distances from the point to all mesh cells
    distances = mesh_cells_gdf.geometry.apply(lambda geom: geom.distance(point_of_interest))
    
    # Find the index of the minimum distance
    nearest_idx = distances.idxmin()
    
    # Get the cell ID for results extraction
    mesh_cell_for_results = mesh_cells_gdf.loc[nearest_idx, 'cell_id']
    mesh_name = mesh_cells_gdf.loc[nearest_idx, 'mesh_name']
    
    print(f"\nNearest cell ID: {mesh_cell_for_results}")
    print(f"Distance: {distances[nearest_idx]:.2f} units")
    print(f"Mesh area: {mesh_name}")
    
    # Extract and store results
    all_results = {}
    max_ws_values = []
    
    # Get results for each plan
    for plan_details in sensitivity_plans:
        plan_number = plan_details['plan_number']
        adjustment = plan_details['adjustment']
        shortid = plan_details['shortid']
        
        # Get the results for this plan
        try:
            results_xr = HdfResultsMesh.get_mesh_cells_timeseries(plan_number)
            
            # Extract water surface data for the specific cell
            ws_data = results_xr[mesh_name]['Water Surface'].sel(cell_id=int(mesh_cell_for_results))
            
            # Convert to DataFrame for easier handling
            ws_df = pd.DataFrame({
                'time': ws_data.time.values,
                'water_surface': ws_data.values
            })
            
            # Store in results dictionary
            all_results[plan_number] = {
                'adjustment': adjustment,
                'shortid': shortid,
                'df': ws_df,
                'max_water_surface': ws_df['water_surface'].max()
            }
            
            # Store the maximum water surface value for summary
            max_ws = ws_df['water_surface'].max()
            max_ws_values.append({
                'plan_number': plan_number,
                'shortid': shortid,
                'adjustment': adjustment,
                'max_water_surface': max_ws
            })
            
            print(f"Plan {plan_number} ({shortid}, adjustment {adjustment:+.3f}): Max Water Surface = {max_ws:.2f}")
            
            # Save individual time series to CSV
            plan_csv_path = results_dir / f"timeseries_{shortid}.csv"
            ws_df.to_csv(plan_csv_path, index=False)
            
        except Exception as e:
            print(f"Error extracting results for plan {plan_number}: {str(e)}")
    
    # Create a summary DataFrame for maximum water surface values
    max_ws_df = pd.DataFrame(max_ws_values)
    max_ws_df = max_ws_df.sort_values('adjustment')
    
    # Save the summary to CSV
    csv_path = results_dir / "manning_sensitivity_bulk_summary.csv"
    max_ws_df.to_csv(csv_path, index=False)
    print(f"\nSummary saved to: {csv_path}")
    
    # Plot the time series results
    plt.figure(figsize=(14, 8))
    
    # Color map for different adjustments (blue for negative, red for positive)
    cmap = plt.cm.coolwarm
    adjustment_range = max(abs(max_ws_df['adjustment'].min()), abs(max_ws_df['adjustment'].max()))
    norm = plt.Normalize(-adjustment_range, adjustment_range)
    
    # Plot time series for each plan
    for _, row in max_ws_df.iterrows():
        plan_number = row['plan_number']
        adjustment = row['adjustment']
        shortid = row['shortid']
        
        # Get the result data
        result_data = all_results[plan_number]
        df = result_data['df']
        
        color = cmap(norm(adjustment))
        label = f"{shortid} ({adjustment:+.3f})"
        
        plt.plot(df['time'], df['water_surface'], label=label, color=color, 
                 linewidth=2 if adjustment == 0 else 1)
    
    # Add plot details
    plt.title(f'Water Surface Sensitivity to Manning\'s Roughness at Cell ID: {mesh_cell_for_results}')
    plt.xlabel('Time')
    plt.ylabel('Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Save the plot
    plot_path = results_dir / "manning_sensitivity_bulk_timeseries.png"
    plt.tight_layout()
    plt.savefig(plot_path)
    print(f"Time series plot saved to: {plot_path}")
    
    # Create a summary plot of max water surface vs adjustment
    plt.figure(figsize=(10, 6))
    plt.scatter(max_ws_df['adjustment'], max_ws_df['max_water_surface'], c=max_ws_df['adjustment'], 
                cmap='coolwarm', norm=norm, s=100)
    
    # Add a trend line
    z = np.polyfit(max_ws_df['adjustment'], max_ws_df['max_water_surface'], 1)
    p = np.poly1d(z)
    plt.plot(max_ws_df['adjustment'], p(max_ws_df['adjustment']), "r--", alpha=0.8)
    
    # Calculate the slope for annotation
    slope = z[0]
    
    # Add plot details
    plt.title(f'Maximum Water Surface vs Manning\'s Roughness Adjustment')
    plt.xlabel('Manning\'s Roughness Adjustment')
    plt.ylabel('Maximum Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add annotation with sensitivity information
    plt.annotate(f'Sensitivity: {slope:.4f} ft per unit n\n(or {slope*0.01:.4f} ft per 0.01 change in n)',
                xy=(0.05, 0.05), xycoords='axes fraction',
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
    
    # Save the summary plot
    summary_plot_path = results_dir / "manning_sensitivity_bulk_summary.png"
    plt.tight_layout()
    plt.savefig(summary_plot_path)
    print(f"Summary plot saved to: {summary_plot_path}")
    
    # Show the plots
    plt.show()
    
    return {
        'all_results': all_results,
        'max_ws_summary': max_ws_df,
        'mesh_cell_id': mesh_cell_for_results,
        'mesh_name': mesh_name
    }
```

```python
def manning_sensitivity_baseoverride(
    template_plan,
    point_of_interest,
    interval=0.005,
    num_intervals=4,
    run_parallel=True,
    max_workers=2,
    num_cores=2,
    results_folder="mannings_sensitivity_results"
):
    """
    Perform a sensitivity analysis by adjusting only the base Manning's roughness values.
    Regional overrides remain unchanged.
    
    Parameters:
    -----------
    template_plan : str
        The plan number to use as a template (e.g., "01")
    point_of_interest : tuple or Point
        Coordinates (x, y) or Point object for finding the nearest mesh cell
    interval : float, optional
        The increment for adjusting Manning's roughness. Default is 0.005.
    num_intervals : int, optional
        Number of intervals to test in each direction (+/-). Default is 4.
    run_parallel : bool, optional
        Whether to run the plans in parallel. Default is True.
    max_workers : int, optional
        Maximum number of parallel workers. Default is 2.
    num_cores : int, optional
        Number of cores per worker. Default is 2.
    results_folder : str, optional
        Name of the folder to save results. Default is "mannings_sensitivity_results".
    
    Returns:
    --------
    dict
        Dictionary containing the sensitivity analysis results
    """
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path
    from shapely.geometry import Point
    import os
    import time
    
    # Convert point_of_interest to Point if it's not already
    if not isinstance(point_of_interest, Point):
        point_of_interest = Point(point_of_interest[0], point_of_interest[1])
    
    # Create results directory
    results_dir = Path(ras.project_folder) / results_folder
    results_dir.mkdir(exist_ok=True)
    
    # Get the geometry number associated with the template plan
    template_geom = ras.plan_df.loc[ras.plan_df['plan_number'] == template_plan, 'geometry_number'].values[0]
    print(f"Template geometry number: {template_geom}")
    
    # Get the geometry file path
    geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == template_geom, 'full_path'].values[0]
    
    # Get the original Manning's values
    original_baseoverrides = RasGeo.get_mannings_baseoverrides(geom_path)
    original_regionoverrides = RasGeo.get_mannings_regionoverrides(geom_path)
    
    # Store the base plan details for reference
    base_plan_details = {
        'plan_number': template_plan,
        'geom_number': template_geom,
        'adjustment': 0.0,
        'shortid': f"Original"
    }
    
    # Create a list to store all the new plan details
    sensitivity_plans = [base_plan_details]
    
    # Generate plans with adjusted Manning's values
    adjustments = []
    
    # Add positive adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(i * interval)
    
    # Add negative adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(-i * interval)
    
    # Sort adjustments
    adjustments.sort()
    
    # Create plans for each adjustment
    for adjustment in adjustments:
        # Format the adjustment string (e.g., P005 for +0.005, N005 for -0.005)
        if adjustment > 0:
            adj_str = f"P{int(adjustment*1000):03d}"
        else:
            adj_str = f"N{int(abs(adjustment)*1000):03d}"
        
        # Create a new plan with adjusted values
        print(f"\nCreating plan with {adjustment:+.3f} Base Manning's adjustment (BO_{adj_str})...")
        
        # Clone the template plan and geometry
        new_plan_number = RasPlan.clone_plan(template_plan, new_plan_shortid=f"BO_{adj_str}")
        new_geom_number = RasPlan.clone_geom(template_geom)
        
        # Set the new plan to use the new geometry
        RasPlan.set_geom(new_plan_number, new_geom_number)
        
        # Get the new geometry file path
        new_geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == new_geom_number, 'full_path'].values[0]
        
        # Create adjusted Base Manning's values (only modify base overrides, not regional)
        modified_baseoverrides = original_baseoverrides.copy()
        modified_baseoverrides["Base Manning's n Value"] = modified_baseoverrides["Base Manning's n Value"] + adjustment
        
        # Apply base Manning's adjustments
        RasGeo.set_mannings_baseoverrides(new_geom_path, modified_baseoverrides)
        
        # Copy original region overrides unchanged (if they exist)
        if not original_regionoverrides.empty:
            RasGeo.set_mannings_regionoverrides(new_geom_path, original_regionoverrides)
        
        # Store the plan details
        sensitivity_plans.append({
            'plan_number': new_plan_number,
            'geom_number': new_geom_number,
            'adjustment': adjustment,
            'shortid': f"BO_{adj_str}"
        })
    
    # Get all plan numbers for execution (excluding the template which is already computed)
    plan_numbers = [plan_details['plan_number'] for plan_details in sensitivity_plans 
                   if plan_details['plan_number'] != template_plan]
    
    # Run the plans (either in parallel or sequentially)
    if run_parallel:
        print(f"\nRunning {len(plan_numbers)} plans in parallel...")
        results = RasCmdr.compute_parallel(
            plan_number=plan_numbers,
            max_workers=max_workers,
            num_cores=num_cores,
            clear_geompre=True
        )
    else:
        print(f"\nRunning {len(plan_numbers)} plans sequentially...")
        results = {}
        for plan_number in plan_numbers:
            print(f"  Running plan {plan_number}...")
            result = RasCmdr.compute_plan(plan_number, clear_geompre=True)
            results[plan_number] = result
    
    print("\nExecution results:")
    for plan, success in results.items():
        print(f"  Plan {plan}: {'Successful' if success else 'Failed'}")
    
    # Use template geometry for mesh cell identification
    geom_number = template_geom
    geom_hdf_path = ras.geom_df.loc[ras.geom_df['geom_number'] == geom_number, 'hdf_path'].values[0]
    
    # Find the nearest mesh cell to the point of interest
    mesh_cells_gdf = HdfMesh.get_mesh_cell_points(geom_hdf_path)
    
    # Calculate distances from the point to all mesh cells
    distances = mesh_cells_gdf.geometry.apply(lambda geom: geom.distance(point_of_interest))
    
    # Find the index of the minimum distance
    nearest_idx = distances.idxmin()
    
    # Get the cell ID for results extraction
    mesh_cell_for_results = mesh_cells_gdf.loc[nearest_idx, 'cell_id']
    mesh_name = mesh_cells_gdf.loc[nearest_idx, 'mesh_name']
    
    print(f"\nNearest cell ID: {mesh_cell_for_results}")
    print(f"Distance: {distances[nearest_idx]:.2f} units")
    print(f"Mesh area: {mesh_name}")
    
    # Extract and store results
    all_results = {}
    max_ws_values = []
    
    # Get results for each plan
    for plan_details in sensitivity_plans:
        plan_number = plan_details['plan_number']
        adjustment = plan_details['adjustment']
        shortid = plan_details['shortid']
        
        # Get the results for this plan
        try:
            results_xr = HdfResultsMesh.get_mesh_cells_timeseries(plan_number)
            
            # Extract water surface data for the specific cell
            ws_data = results_xr[mesh_name]['Water Surface'].sel(cell_id=int(mesh_cell_for_results))
            
            # Convert to DataFrame for easier handling
            ws_df = pd.DataFrame({
                'time': ws_data.time.values,
                'water_surface': ws_data.values
            })
            
            # Store in results dictionary
            all_results[plan_number] = {
                'adjustment': adjustment,
                'shortid': shortid,
                'df': ws_df,
                'max_water_surface': ws_df['water_surface'].max()
            }
            
            # Store the maximum water surface value for summary
            max_ws = ws_df['water_surface'].max()
            max_ws_values.append({
                'plan_number': plan_number,
                'shortid': shortid,
                'adjustment': adjustment,
                'max_water_surface': max_ws
            })
            
            print(f"Plan {plan_number} ({shortid}, adjustment {adjustment:+.3f}): Max Water Surface = {max_ws:.2f}")
            
            # Save individual time series to CSV
            plan_csv_path = results_dir / f"timeseries_{shortid}.csv"
            ws_df.to_csv(plan_csv_path, index=False)
            
        except Exception as e:
            print(f"Error extracting results for plan {plan_number}: {str(e)}")
    
    # Create a summary DataFrame for maximum water surface values
    max_ws_df = pd.DataFrame(max_ws_values)
    max_ws_df = max_ws_df.sort_values('adjustment')
    
    # Save the summary to CSV
    csv_path = results_dir / "manning_sensitivity_baseoverride_summary.csv"
    max_ws_df.to_csv(csv_path, index=False)
    print(f"\nSummary saved to: {csv_path}")
    
    # Plot the time series results
    plt.figure(figsize=(14, 8))
    
    # Color map for different adjustments (blue for negative, red for positive)
    cmap = plt.cm.coolwarm
    adjustment_range = max(abs(max_ws_df['adjustment'].min()), abs(max_ws_df['adjustment'].max()))
    norm = plt.Normalize(-adjustment_range, adjustment_range)
    
    # Plot time series for each plan
    for _, row in max_ws_df.iterrows():
        plan_number = row['plan_number']
        adjustment = row['adjustment']
        shortid = row['shortid']
        
        # Get the result data
        result_data = all_results[plan_number]
        df = result_data['df']
        
        color = cmap(norm(adjustment))
        label = f"{shortid} ({adjustment:+.3f})"
        
        plt.plot(df['time'], df['water_surface'], label=label, color=color, 
                 linewidth=2 if adjustment == 0 else 1)
    
    # Add plot details
    plt.title(f'Water Surface Sensitivity to Base Manning\'s Roughness at Cell ID: {mesh_cell_for_results}\n(Regional Overrides Unchanged)')
    plt.xlabel('Time')
    plt.ylabel('Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Save the plot
    plot_path = results_dir / "manning_sensitivity_baseoverride_timeseries.png"
    plt.tight_layout()
    plt.savefig(plot_path)
    print(f"Time series plot saved to: {plot_path}")
    
    # Create a summary plot of max water surface vs adjustment
    plt.figure(figsize=(10, 6))
    plt.scatter(max_ws_df['adjustment'], max_ws_df['max_water_surface'], c=max_ws_df['adjustment'], 
                cmap='coolwarm', norm=norm, s=100)
    
    # Add a trend line
    z = np.polyfit(max_ws_df['adjustment'], max_ws_df['max_water_surface'], 1)
    p = np.poly1d(z)
    plt.plot(max_ws_df['adjustment'], p(max_ws_df['adjustment']), "r--", alpha=0.8)
    
    # Calculate the slope for annotation
    slope = z[0]
    
    # Add plot details
    plt.title(f'Maximum Water Surface vs Base Manning\'s Roughness Adjustment\n(Regional Overrides Unchanged)')
    plt.xlabel('Base Manning\'s Roughness Adjustment')
    plt.ylabel('Maximum Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add annotation with sensitivity information
    plt.annotate(f'Sensitivity: {slope:.4f} ft per unit n\n(or {slope*0.01:.4f} ft per 0.01 change in n)',
                xy=(0.05, 0.05), xycoords='axes fraction',
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
    
    # Save the summary plot
    summary_plot_path = results_dir / "manning_sensitivity_baseoverride_summary.png"
    plt.tight_layout()
    plt.savefig(summary_plot_path)
    print(f"Summary plot saved to: {summary_plot_path}")
    
    # Show the plots
    plt.show()
    
    return {
        'all_results': all_results,
        'max_ws_summary': max_ws_df,
        'mesh_cell_id': mesh_cell_for_results,
        'mesh_name': mesh_name
    }
```

```python
def manning_sensitivity_region(
    template_plan,
    point_of_interest,
    table_num=None,
    interval=0.005,
    num_intervals=4,
    run_parallel=True,
    max_workers=2,
    num_cores=2,
    results_folder="mannings_sensitivity_results"
):
    """
    Perform a sensitivity analysis by adjusting only regional Manning's roughness values.
    Base overrides remain unchanged.
    
    Parameters:
    -----------
    template_plan : str
        The plan number to use as a template (e.g., "01")
    point_of_interest : tuple or Point
        Coordinates (x, y) or Point object for finding the nearest mesh cell
    table_num : str or None, optional
        Specific region table number to adjust. If None, all regions are adjusted.
    interval : float, optional
        The increment for adjusting Manning's roughness. Default is 0.005.
    num_intervals : int, optional
        Number of intervals to test in each direction (+/-). Default is 4.
    run_parallel : bool, optional
        Whether to run the plans in parallel. Default is True.
    max_workers : int, optional
        Maximum number of parallel workers. Default is 2.
    num_cores : int, optional
        Number of cores per worker. Default is 2.
    results_folder : str, optional
        Name of the folder to save results. Default is "mannings_sensitivity_results".
    
    Returns:
    --------
    dict
        Dictionary containing the sensitivity analysis results
    """
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path
    from shapely.geometry import Point
    import os
    import time
    
    # Convert point_of_interest to Point if it's not already
    if not isinstance(point_of_interest, Point):
        point_of_interest = Point(point_of_interest[0], point_of_interest[1])
    
    # Create results directory
    results_dir = Path(ras.project_folder) / results_folder
    results_dir.mkdir(exist_ok=True)
    
    # Get the geometry number associated with the template plan
    template_geom = ras.plan_df.loc[ras.plan_df['plan_number'] == template_plan, 'geometry_number'].values[0]
    print(f"Template geometry number: {template_geom}")
    
    # Get the geometry file path
    geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == template_geom, 'full_path'].values[0]
    
    # Get the original Manning's values
    original_baseoverrides = RasGeo.get_mannings_baseoverrides(geom_path)
    original_regionoverrides = RasGeo.get_mannings_regionoverrides(geom_path)
    
    # Verify that we have region overrides to work with
    if original_regionoverrides.empty:
        print("No regional Manning's overrides found in the model. Cannot perform region sensitivity analysis.")
        return None
        
    # If table_num is specified, verify it exists in the region overrides
    if table_num is not None:
        if str(table_num) not in original_regionoverrides['Table Number'].unique():
            print(f"Table number {table_num} not found in regional overrides. Available tables: {original_regionoverrides['Table Number'].unique()}")
            return None
        else:
            print(f"Adjusting only region table {table_num}")
            # Filter to the specified table
            region_tables_to_adjust = [str(table_num)]
            # Generate shortid prefix
            shortid_prefix = f"RG_T{table_num}_"
    else:
        # Adjust all region tables
        region_tables_to_adjust = original_regionoverrides['Table Number'].unique()
        shortid_prefix = "RG_"
        print(f"Adjusting all region tables: {region_tables_to_adjust}")
    
    # Store the base plan details for reference
    base_plan_details = {
        'plan_number': template_plan,
        'geom_number': template_geom,
        'adjustment': 0.0,
        'shortid': f"Original"
    }
    
    # Create a list to store all the new plan details
    sensitivity_plans = [base_plan_details]
    
    # Generate plans with adjusted Manning's values
    adjustments = []
    
    # Add positive adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(i * interval)
    
    # Add negative adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(-i * interval)
    
    # Sort adjustments
    adjustments.sort()
    
    # Create plans for each adjustment
    for adjustment in adjustments:
        # Format the adjustment string (e.g., P005 for +0.005, N005 for -0.005)
        if adjustment > 0:
            adj_str = f"P{int(adjustment*1000):03d}"
        else:
            adj_str = f"N{int(abs(adjustment)*1000):03d}"
        
        # Create a new plan with adjusted values
        shortid = f"{shortid_prefix}{adj_str}"
        print(f"\nCreating plan with {adjustment:+.3f} Region Manning's adjustment ({shortid})...")
        
        # Clone the template plan and geometry
        new_plan_number = RasPlan.clone_plan(template_plan, new_plan_shortid=shortid)
        new_geom_number = RasPlan.clone_geom(template_geom)
        
        # Set the new plan to use the new geometry
        RasPlan.set_geom(new_plan_number, new_geom_number)
        
        # Get the new geometry file path
        new_geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == new_geom_number, 'full_path'].values[0]
        
        # Copy original base overrides unchanged
        RasGeo.set_mannings_baseoverrides(new_geom_path, original_baseoverrides)
        
        # Create adjusted Region Manning's values
        modified_regionoverrides = original_regionoverrides.copy()
        
        # Only adjust the specified table numbers
        for table in region_tables_to_adjust:
            # Create a mask for rows with the current table number
            mask = modified_regionoverrides['Table Number'] == table
            # Apply the adjustment to those rows
            modified_regionoverrides.loc[mask, 'MainChannel'] = modified_regionoverrides.loc[mask, 'MainChannel'] + adjustment
        
        # Apply region Manning's adjustments
        RasGeo.set_mannings_regionoverrides(new_geom_path, modified_regionoverrides)
        
        # Store the plan details
        sensitivity_plans.append({
            'plan_number': new_plan_number,
            'geom_number': new_geom_number,
            'adjustment': adjustment,
            'shortid': shortid
        })
    
    # Get all plan numbers for execution (excluding the template which is already computed)
    plan_numbers = [plan_details['plan_number'] for plan_details in sensitivity_plans 
                   if plan_details['plan_number'] != template_plan]
    
    # Run the plans (either in parallel or sequentially)
    if run_parallel:
        print(f"\nRunning {len(plan_numbers)} plans in parallel...")
        results = RasCmdr.compute_parallel(
            plan_number=plan_numbers,
            max_workers=max_workers,
            num_cores=num_cores,
            clear_geompre=True
        )
    else:
        print(f"\nRunning {len(plan_numbers)} plans sequentially...")
        results = {}
        for plan_number in plan_numbers:
            print(f"  Running plan {plan_number}...")
            result = RasCmdr.compute_plan(plan_number, clear_geompre=True)
            results[plan_number] = result
    
    print("\nExecution results:")
    for plan, success in results.items():
        print(f"  Plan {plan}: {'Successful' if success else 'Failed'}")
    
    # Use template geometry for mesh cell identification
    geom_number = template_geom
    geom_hdf_path = ras.geom_df.loc[ras.geom_df['geom_number'] == geom_number, 'hdf_path'].values[0]
    
    # Find the nearest mesh cell to the point of interest
    mesh_cells_gdf = HdfMesh.get_mesh_cell_points(geom_hdf_path)
    
    # Calculate distances from the point to all mesh cells
    distances = mesh_cells_gdf.geometry.apply(lambda geom: geom.distance(point_of_interest))
    
    # Find the index of the minimum distance
    nearest_idx = distances.idxmin()
    
    # Get the cell ID for results extraction
    mesh_cell_for_results = mesh_cells_gdf.loc[nearest_idx, 'cell_id']
    mesh_name = mesh_cells_gdf.loc[nearest_idx, 'mesh_name']
    
    print(f"\nNearest cell ID: {mesh_cell_for_results}")
    print(f"Distance: {distances[nearest_idx]:.2f} units")
    print(f"Mesh area: {mesh_name}")
    
    # Extract and store results
    all_results = {}
    max_ws_values = []
    
    # Get results for each plan
    for plan_details in sensitivity_plans:
        plan_number = plan_details['plan_number']
        adjustment = plan_details['adjustment']
        shortid = plan_details['shortid']
        
        # Get the results for this plan
        try:
            results_xr = HdfResultsMesh.get_mesh_cells_timeseries(plan_number)
            
            # Extract water surface data for the specific cell
            ws_data = results_xr[mesh_name]['Water Surface'].sel(cell_id=int(mesh_cell_for_results))
            
            # Convert to DataFrame for easier handling
            ws_df = pd.DataFrame({
                'time': ws_data.time.values,
                'water_surface': ws_data.values
            })
            
            # Store in results dictionary
            all_results[plan_number] = {
                'adjustment': adjustment,
                'shortid': shortid,
                'df': ws_df,
                'max_water_surface': ws_df['water_surface'].max()
            }
            
            # Store the maximum water surface value for summary
            max_ws = ws_df['water_surface'].max()
            max_ws_values.append({
                'plan_number': plan_number,
                'shortid': shortid,
                'adjustment': adjustment,
                'max_water_surface': max_ws
            })
            
            print(f"Plan {plan_number} ({shortid}, adjustment {adjustment:+.3f}): Max Water Surface = {max_ws:.2f}")
            
            # Save individual time series to CSV
            plan_csv_path = results_dir / f"timeseries_{shortid}.csv"
            ws_df.to_csv(plan_csv_path, index=False)
            
        except Exception as e:
            print(f"Error extracting results for plan {plan_number}: {str(e)}")
    
    # Create a summary DataFrame for maximum water surface values
    max_ws_df = pd.DataFrame(max_ws_values)
    max_ws_df = max_ws_df.sort_values('adjustment')
    
    # Save the summary to CSV
    file_prefix = "manning_sensitivity_region"
    if table_num is not None:
        file_prefix += f"_table{table_num}"
    
    csv_path = results_dir / f"{file_prefix}_summary.csv"
    max_ws_df.to_csv(csv_path, index=False)
    print(f"\nSummary saved to: {csv_path}")
    
    # Plot the time series results
    plt.figure(figsize=(14, 8))
    
    # Color map for different adjustments (blue for negative, red for positive)
    cmap = plt.cm.coolwarm
    adjustment_range = max(abs(max_ws_df['adjustment'].min()), abs(max_ws_df['adjustment'].max()))
    norm = plt.Normalize(-adjustment_range, adjustment_range)
    
    # Plot time series for each plan
    for _, row in max_ws_df.iterrows():
        plan_number = row['plan_number']
        adjustment = row['adjustment']
        shortid = row['shortid']
        
        # Get the result data
        result_data = all_results[plan_number]
        df = result_data['df']
        
        color = cmap(norm(adjustment))
        label = f"{shortid} ({adjustment:+.3f})"
        
        plt.plot(df['time'], df['water_surface'], label=label, color=color, 
                 linewidth=2 if adjustment == 0 else 1)
    
    # Add plot details
    region_info = f"Region Table {table_num}" if table_num is not None else "All Regions"
    plt.title(f'Water Surface Sensitivity to Region Manning\'s Roughness ({region_info}) at Cell ID: {mesh_cell_for_results}\n(Base Overrides Unchanged)')
    plt.xlabel('Time')
    plt.ylabel('Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Save the plot
    plot_path = results_dir / f"{file_prefix}_timeseries.png"
    plt.tight_layout()
    plt.savefig(plot_path)
    print(f"Time series plot saved to: {plot_path}")
    
    # Create a summary plot of max water surface vs adjustment
    plt.figure(figsize=(10, 6))
    plt.scatter(max_ws_df['adjustment'], max_ws_df['max_water_surface'], c=max_ws_df['adjustment'], 
                cmap='coolwarm', norm=norm, s=100)
    
    # Add a trend line
    z = np.polyfit(max_ws_df['adjustment'], max_ws_df['max_water_surface'], 1)
    p = np.poly1d(z)
    plt.plot(max_ws_df['adjustment'], p(max_ws_df['adjustment']), "r--", alpha=0.8)
    
    # Calculate the slope for annotation
    slope = z[0]
    
    # Add plot details
    plt.title(f'Maximum Water Surface vs Region Manning\'s Roughness Adjustment ({region_info})\n(Base Overrides Unchanged)')
    plt.xlabel('Region Manning\'s Roughness Adjustment')
    plt.ylabel('Maximum Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add annotation with sensitivity information
    plt.annotate(f'Sensitivity: {slope:.4f} ft per unit n\n(or {slope*0.01:.4f} ft per 0.01 change in n)',
                xy=(0.05, 0.05), xycoords='axes fraction',
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
    
    # Save the summary plot
    summary_plot_path = results_dir / f"{file_prefix}_summary.png"
    plt.tight_layout()
    plt.savefig(summary_plot_path)
    print(f"Summary plot saved to: {summary_plot_path}")
    
    # Show the plots
    plt.show()
    
    return {
        'all_results': all_results,
        'max_ws_summary': max_ws_df,
        'mesh_cell_id': mesh_cell_for_results,
        'mesh_name': mesh_name,
        'region_tables': region_tables_to_adjust
    }
```

```python
def manning_sensitivity_base_landuse(
    template_plan,
    point_of_interest,
    landuse_types,
    interval=0.005,
    num_intervals=4,
    run_parallel=True,
    max_workers=2,
    num_cores=2,
    results_folder="mannings_sensitivity_results"
):
    """
    Perform a sensitivity analysis by adjusting specific land use types in base Manning's roughness values.
    All other land uses and regional overrides remain unchanged.
    
    Parameters:
    -----------
    template_plan : str
        The plan number to use as a template (e.g., "01")
    point_of_interest : tuple or Point
        Coordinates (x, y) or Point object for finding the nearest mesh cell
    landuse_types : str or list
        Land use type(s) to adjust (must match names in the Manning's table)
    interval : float, optional
        The increment for adjusting Manning's roughness. Default is 0.005.
    num_intervals : int, optional
        Number of intervals to test in each direction (+/-). Default is 4.
    run_parallel : bool, optional
        Whether to run the plans in parallel. Default is True.
    max_workers : int, optional
        Maximum number of parallel workers. Default is 2.
    num_cores : int, optional
        Number of cores per worker. Default is 2.
    results_folder : str, optional
        Name of the folder to save results. Default is "mannings_sensitivity_results".
    
    Returns:
    --------
    dict
        Dictionary containing the sensitivity analysis results
    """
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path
    from shapely.geometry import Point
    import os
    import time
    import re
    
    # Convert point_of_interest to Point if it's not already
    if not isinstance(point_of_interest, Point):
        point_of_interest = Point(point_of_interest[0], point_of_interest[1])
    
    # Convert landuse_types to list if it's a string
    if isinstance(landuse_types, str):
        landuse_types = [landuse_types]
    
    # Create results directory
    results_dir = Path(ras.project_folder) / results_folder
    results_dir.mkdir(exist_ok=True)
    
    # Get the geometry number associated with the template plan
    template_geom = ras.plan_df.loc[ras.plan_df['plan_number'] == template_plan, 'geometry_number'].values[0]
    print(f"Template geometry number: {template_geom}")
    
    # Get the geometry file path
    geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == template_geom, 'full_path'].values[0]
    
    # Get the original Manning's values
    original_baseoverrides = RasGeo.get_mannings_baseoverrides(geom_path)
    original_regionoverrides = RasGeo.get_mannings_regionoverrides(geom_path)
    
    # Check if the specified land use types exist in the base overrides
    missing_landuses = [lu for lu in landuse_types if lu not in original_baseoverrides['Land Cover Name'].values]
    if missing_landuses:
        print(f"The following land use types were not found in the base Manning's overrides: {missing_landuses}")
        available_landuses = original_baseoverrides['Land Cover Name'].unique()
        print(f"Available land use types: {available_landuses}")
        return None
    
    # Create abbreviated codes for land use types (for use in ShortIDs)
    landuse_codes = {}
    for lu in landuse_types:
        # Create an abbreviation using first letters of words or first 2-3 chars
        if ' ' in lu:
            # For multi-word land uses, use first letter of each word
            code = ''.join(word[0] for word in lu.split())
        else:
            # For single word, use first 2 characters
            code = lu[:2].upper()
        landuse_codes[lu] = code
    
    print(f"Land use types to adjust: {landuse_types}")
    print(f"Land use codes for ShortIDs: {landuse_codes}")
    
    # Create ShortID suffix based on land use codes
    landuse_shortid = '_'.join(landuse_codes.values())
    
    # Store the base plan details for reference
    base_plan_details = {
        'plan_number': template_plan,
        'geom_number': template_geom,
        'adjustment': 0.0,
        'shortid': f"Original",
        'landuse_types': landuse_types
    }
    
    # Create a list to store all the new plan details
    sensitivity_plans = [base_plan_details]
    
    # Generate plans with adjusted Manning's values
    adjustments = []
    
    # Add positive adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(i * interval)
    
    # Add negative adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(-i * interval)
    
    # Sort adjustments
    adjustments.sort()
    
    # Create plans for each adjustment
    for adjustment in adjustments:
        # Format the adjustment string (e.g., P005 for +0.005, N005 for -0.005)
        if adjustment > 0:
            adj_str = f"P{int(adjustment*1000):03d}"
        else:
            adj_str = f"N{int(abs(adjustment)*1000):03d}"
        
        # Create a new plan with adjusted values
        shortid = f"BL_{landuse_shortid}_{adj_str}"
        print(f"\nCreating plan with {adjustment:+.3f} adjustment to land uses ({shortid})...")
        
        # Clone the template plan and geometry
        new_plan_number = RasPlan.clone_plan(template_plan, new_plan_shortid=shortid)
        new_geom_number = RasPlan.clone_geom(template_geom)
        
        # Set the new plan to use the new geometry
        RasPlan.set_geom(new_plan_number, new_geom_number)
        
        # Get the new geometry file path
        new_geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == new_geom_number, 'full_path'].values[0]
        
        # Create adjusted base Manning's values (only modify specified land uses)
        modified_baseoverrides = original_baseoverrides.copy()
        
        # Apply adjustments only to the specified land use types
        for landuse in landuse_types:
            mask = modified_baseoverrides['Land Cover Name'] == landuse
            if mask.any():
                current_value = modified_baseoverrides.loc[mask, "Base Manning's n Value"].values[0]
                print(f"  Adjusting '{landuse}' from {current_value:.4f} to {current_value + adjustment:.4f}")
                modified_baseoverrides.loc[mask, "Base Manning's n Value"] += adjustment
        
        # Apply base Manning's adjustments
        RasGeo.set_mannings_baseoverrides(new_geom_path, modified_baseoverrides)
        
        # Copy original region overrides unchanged (if they exist)
        if not original_regionoverrides.empty:
            RasGeo.set_mannings_regionoverrides(new_geom_path, original_regionoverrides)
        
        # Store the plan details
        sensitivity_plans.append({
            'plan_number': new_plan_number,
            'geom_number': new_geom_number,
            'adjustment': adjustment,
            'shortid': shortid,
            'landuse_types': landuse_types
        })
    
    # Get all plan numbers for execution (excluding the template which is already computed)
    plan_numbers = [plan_details['plan_number'] for plan_details in sensitivity_plans 
                   if plan_details['plan_number'] != template_plan]
    
    # Run the plans (either in parallel or sequentially)
    if run_parallel:
        print(f"\nRunning {len(plan_numbers)} plans in parallel...")
        results = RasCmdr.compute_parallel(
            plan_number=plan_numbers,
            max_workers=max_workers,
            num_cores=num_cores,
            clear_geompre=True
        )
    else:
        print(f"\nRunning {len(plan_numbers)} plans sequentially...")
        results = {}
        for plan_number in plan_numbers:
            print(f"  Running plan {plan_number}...")
            result = RasCmdr.compute_plan(plan_number, clear_geompre=True)
            results[plan_number] = result
    
    print("\nExecution results:")
    for plan, success in results.items():
        print(f"  Plan {plan}: {'Successful' if success else 'Failed'}")
    
    # Use template geometry for mesh cell identification
    geom_number = template_geom
    geom_hdf_path = ras.geom_df.loc[ras.geom_df['geom_number'] == geom_number, 'hdf_path'].values[0]
    
    # Find the nearest mesh cell to the point of interest
    mesh_cells_gdf = HdfMesh.get_mesh_cell_points(geom_hdf_path)
    
    # Calculate distances from the point to all mesh cells
    distances = mesh_cells_gdf.geometry.apply(lambda geom: geom.distance(point_of_interest))
    
    # Find the index of the minimum distance
    nearest_idx = distances.idxmin()
    
    # Get the cell ID for results extraction
    mesh_cell_for_results = mesh_cells_gdf.loc[nearest_idx, 'cell_id']
    mesh_name = mesh_cells_gdf.loc[nearest_idx, 'mesh_name']
    
    print(f"\nNearest cell ID: {mesh_cell_for_results}")
    print(f"Distance: {distances[nearest_idx]:.2f} units")
    print(f"Mesh area: {mesh_name}")
    
    # Extract and store results
    all_results = {}
    max_ws_values = []
    
    # Get results for each plan
    for plan_details in sensitivity_plans:
        plan_number = plan_details['plan_number']
        adjustment = plan_details['adjustment']
        shortid = plan_details['shortid']
        
        # Get the results for this plan
        try:
            results_xr = HdfResultsMesh.get_mesh_cells_timeseries(plan_number)
            
            # Extract water surface data for the specific cell
            ws_data = results_xr[mesh_name]['Water Surface'].sel(cell_id=int(mesh_cell_for_results))
            
            # Convert to DataFrame for easier handling
            ws_df = pd.DataFrame({
                'time': ws_data.time.values,
                'water_surface': ws_data.values
            })
            
            # Store in results dictionary
            all_results[plan_number] = {
                'adjustment': adjustment,
                'shortid': shortid,
                'df': ws_df,
                'max_water_surface': ws_df['water_surface'].max()
            }
            
            # Store the maximum water surface value for summary
            max_ws = ws_df['water_surface'].max()
            max_ws_values.append({
                'plan_number': plan_number,
                'shortid': shortid,
                'adjustment': adjustment,
                'max_water_surface': max_ws
            })
            
            print(f"Plan {plan_number} ({shortid}, adjustment {adjustment:+.3f}): Max Water Surface = {max_ws:.2f}")
            
            # Save individual time series to CSV
            plan_csv_path = results_dir / f"timeseries_{shortid}.csv"
            ws_df.to_csv(plan_csv_path, index=False)
            
        except Exception as e:
            print(f"Error extracting results for plan {plan_number}: {str(e)}")
    
    # Create a summary DataFrame for maximum water surface values
    max_ws_df = pd.DataFrame(max_ws_values)
    max_ws_df = max_ws_df.sort_values('adjustment')
    
    # Create a friendly land use description for plot titles
    landuse_desc = ", ".join(landuse_types)
    if len(landuse_desc) > 40:  # Truncate if too long
        landuse_desc = landuse_desc[:37] + "..."
    
    # Save the summary to CSV
    csv_path = results_dir / f"manning_sensitivity_base_landuse_{landuse_shortid}_summary.csv"
    max_ws_df.to_csv(csv_path, index=False)
    print(f"\nSummary saved to: {csv_path}")
    
    # Plot the time series results
    plt.figure(figsize=(14, 8))
    
    # Color map for different adjustments (blue for negative, red for positive)
    cmap = plt.cm.coolwarm
    adjustment_range = max(abs(max_ws_df['adjustment'].min()), abs(max_ws_df['adjustment'].max()))
    norm = plt.Normalize(-adjustment_range, adjustment_range)
    
    # Plot time series for each plan
    for _, row in max_ws_df.iterrows():
        plan_number = row['plan_number']
        adjustment = row['adjustment']
        shortid = row['shortid']
        
        # Get the result data
        result_data = all_results[plan_number]
        df = result_data['df']
        
        color = cmap(norm(adjustment))
        label = f"{shortid} ({adjustment:+.3f})"
        
        plt.plot(df['time'], df['water_surface'], label=label, color=color, 
                 linewidth=2 if adjustment == 0 else 1)
    
    # Add plot details
    plt.title(f'Water Surface Sensitivity to Base Manning\'s Roughness for Land Uses:\n{landuse_desc}\nCell ID: {mesh_cell_for_results}')
    plt.xlabel('Time')
    plt.ylabel('Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Save the plot
    plot_path = results_dir / f"manning_sensitivity_base_landuse_{landuse_shortid}_timeseries.png"
    plt.tight_layout()
    plt.savefig(plot_path)
    print(f"Time series plot saved to: {plot_path}")
    
    # Create a summary plot of max water surface vs adjustment
    plt.figure(figsize=(10, 6))
    plt.scatter(max_ws_df['adjustment'], max_ws_df['max_water_surface'], c=max_ws_df['adjustment'], 
                cmap='coolwarm', norm=norm, s=100)
    
    # Add a trend line
    z = np.polyfit(max_ws_df['adjustment'], max_ws_df['max_water_surface'], 1)
    p = np.poly1d(z)
    plt.plot(max_ws_df['adjustment'], p(max_ws_df['adjustment']), "r--", alpha=0.8)
    
    # Calculate the slope for annotation
    slope = z[0]
    
    # Add plot details
    plt.title(f'Maximum Water Surface vs Base Manning\'s Roughness Adjustment\nfor Land Uses: {landuse_desc}')
    plt.xlabel('Manning\'s Roughness Adjustment')
    plt.ylabel('Maximum Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add annotation with sensitivity information
    plt.annotate(f'Sensitivity: {slope:.4f} ft per unit n\n(or {slope*0.01:.4f} ft per 0.01 change in n)',
                xy=(0.05, 0.05), xycoords='axes fraction',
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
    
    # Save the summary plot
    summary_plot_path = results_dir / f"manning_sensitivity_base_landuse_{landuse_shortid}_summary.png"
    plt.tight_layout()
    plt.savefig(summary_plot_path)
    print(f"Summary plot saved to: {summary_plot_path}")
    
    # Show the plots
    plt.show()
    
    return {
        'all_results': all_results,
        'max_ws_summary': max_ws_df,
        'mesh_cell_id': mesh_cell_for_results,
        'mesh_name': mesh_name,
        'landuse_types': landuse_types,
        'landuse_codes': landuse_codes
    }
```

### manning_sensitivity_region_landuse

```python
def manning_sensitivity_region_landuse(
    template_plan,
    point_of_interest,
    landuse_types,
    table_num=None,
    interval=0.005,
    num_intervals=4,
    run_parallel=True,
    max_workers=2,
    num_cores=2,
    results_folder="mannings_sensitivity_results"
):
    """
    Perform a sensitivity analysis by adjusting specific land use types in regional Manning's roughness values.
    All other land uses and base overrides remain unchanged.
    
    Parameters:
    -----------
    template_plan : str
        The plan number to use as a template (e.g., "01")
    point_of_interest : tuple or Point
        Coordinates (x, y) or Point object for finding the nearest mesh cell
    landuse_types : str or list
        Land use type(s) to adjust (must match names in the Manning's table)
    table_num : str or None, optional
        Specific region table number to adjust. If None, all regions are adjusted.
    interval : float, optional
        The increment for adjusting Manning's roughness. Default is 0.005.
    num_intervals : int, optional
        Number of intervals to test in each direction (+/-). Default is 4.
    run_parallel : bool, optional
        Whether to run the plans in parallel. Default is True.
    max_workers : int, optional
        Maximum number of parallel workers. Default is 2.
    num_cores : int, optional
        Number of cores per worker. Default is 2.
    results_folder : str, optional
        Name of the folder to save results. Default is "mannings_sensitivity_results".
    
    Returns:
    --------
    dict
        Dictionary containing the sensitivity analysis results
    """
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    from pathlib import Path
    from shapely.geometry import Point
    import os
    import time
    import re
    
    # Convert point_of_interest to Point if it's not already
    if not isinstance(point_of_interest, Point):
        point_of_interest = Point(point_of_interest[0], point_of_interest[1])
    
    # Convert landuse_types to list if it's a string
    if isinstance(landuse_types, str):
        landuse_types = [landuse_types]
    
    # Create results directory
    results_dir = Path(ras.project_folder) / results_folder
    results_dir.mkdir(exist_ok=True)
    
    # Get the geometry number associated with the template plan
    template_geom = ras.plan_df.loc[ras.plan_df['plan_number'] == template_plan, 'geometry_number'].values[0]
    print(f"Template geometry number: {template_geom}")
    
    # Get the geometry file path
    geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == template_geom, 'full_path'].values[0]
    
    # Get the original Manning's values
    original_baseoverrides = RasGeo.get_mannings_baseoverrides(geom_path)
    original_regionoverrides = RasGeo.get_mannings_regionoverrides(geom_path)
    
    # Verify that we have region overrides to work with
    if original_regionoverrides.empty:
        print("No regional Manning's overrides found in the model. Cannot perform region land use sensitivity analysis.")
        return None
    
    # If table_num is specified, verify it exists in the region overrides
    if table_num is not None:
        if str(table_num) not in original_regionoverrides['Table Number'].unique():
            print(f"Table number {table_num} not found in regional overrides. Available tables: {original_regionoverrides['Table Number'].unique()}")
            return None
        else:
            print(f"Adjusting land uses only in region table {table_num}")
            # Filter to the specified table
            region_tables_to_adjust = [str(table_num)]
            # Generate table info for shortid
            table_prefix = f"T{table_num}_"
    else:
        # Adjust all region tables
        region_tables_to_adjust = original_regionoverrides['Table Number'].unique()
        table_prefix = ""
        print(f"Adjusting land uses in all region tables: {region_tables_to_adjust}")
    
    # Check if the specified land use types exist in the region overrides for the tables of interest
    landuse_in_region = False
    for table in region_tables_to_adjust:
        table_data = original_regionoverrides[original_regionoverrides['Table Number'] == table]
        for lu in landuse_types:
            if lu in table_data['Land Cover Name'].values:
                landuse_in_region = True
                break
        if landuse_in_region:
            break
    
    if not landuse_in_region:
        print(f"None of the specified land use types {landuse_types} were found in the selected region tables.")
        available_landuses = original_regionoverrides['Land Cover Name'].unique()
        print(f"Available land uses in regions: {available_landuses}")
        return None
    
    # Create abbreviated codes for land use types (for use in ShortIDs)
    landuse_codes = {}
    for lu in landuse_types:
        # Create an abbreviation using first letters of words or first 2-3 chars
        if ' ' in lu:
            # For multi-word land uses, use first letter of each word
            code = ''.join(word[0] for word in lu.split())
        else:
            # For single word, use first 2 characters
            code = lu[:2].upper()
        landuse_codes[lu] = code
    
    print(f"Land use types to adjust: {landuse_types}")
    print(f"Land use codes for ShortIDs: {landuse_codes}")
    
    # Create ShortID suffix based on land use codes
    landuse_shortid = '_'.join(landuse_codes.values())
    
    # Store the base plan details for reference
    base_plan_details = {
        'plan_number': template_plan,
        'geom_number': template_geom,
        'adjustment': 0.0,
        'shortid': f"Original",
        'landuse_types': landuse_types,
        'region_tables': region_tables_to_adjust
    }
    
    # Create a list to store all the new plan details
    sensitivity_plans = [base_plan_details]
    
    # Generate plans with adjusted Manning's values
    adjustments = []
    
    # Add positive adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(i * interval)
    
    # Add negative adjustments
    for i in range(1, num_intervals + 1):
        adjustments.append(-i * interval)
    
    # Sort adjustments
    adjustments.sort()
    
    # Create plans for each adjustment
    for adjustment in adjustments:
        # Format the adjustment string (e.g., P005 for +0.005, N005 for -0.005)
        if adjustment > 0:
            adj_str = f"P{int(adjustment*1000):03d}"
        else:
            adj_str = f"N{int(abs(adjustment)*1000):03d}"
        
        # Create a new plan with adjusted values
        shortid = f"RL_{table_prefix}{landuse_shortid}_{adj_str}"
        print(f"\nCreating plan with {adjustment:+.3f} adjustment to region land uses ({shortid})...")
        
        # Clone the template plan and geometry
        new_plan_number = RasPlan.clone_plan(template_plan, new_plan_shortid=shortid)
        new_geom_number = RasPlan.clone_geom(template_geom)
        
        # Set the new plan to use the new geometry
        RasPlan.set_geom(new_plan_number, new_geom_number)
        
        # Get the new geometry file path
        new_geom_path = ras.geom_df.loc[ras.geom_df['geom_number'] == new_geom_number, 'full_path'].values[0]
        
        # Copy original base overrides unchanged
        RasGeo.set_mannings_baseoverrides(new_geom_path, original_baseoverrides)
        
        # Create adjusted Region Manning's values
        modified_regionoverrides = original_regionoverrides.copy()
        
        # For each table to adjust
        for table in region_tables_to_adjust:
            # For each land use type to adjust
            for landuse in landuse_types:
                # Create a mask for rows with the current table number and land use type
                mask = (modified_regionoverrides['Table Number'] == table) & (modified_regionoverrides['Land Cover Name'] == landuse)
                
                # If this land use exists in this table, adjust it
                if mask.any():
                    current_value = modified_regionoverrides.loc[mask, 'MainChannel'].values[0]
                    print(f"  Adjusting '{landuse}' in table {table} from {current_value:.4f} to {current_value + adjustment:.4f}")
                    modified_regionoverrides.loc[mask, 'MainChannel'] += adjustment
        
        # Apply region Manning's adjustments
        RasGeo.set_mannings_regionoverrides(new_geom_path, modified_regionoverrides)
        
        # Store the plan details
        sensitivity_plans.append({
            'plan_number': new_plan_number,
            'geom_number': new_geom_number,
            'adjustment': adjustment,
            'shortid': shortid,
            'landuse_types': landuse_types,
            'region_tables': region_tables_to_adjust
        })
    
    # Get all plan numbers for execution (excluding the template which is already computed)
    plan_numbers = [plan_details['plan_number'] for plan_details in sensitivity_plans 
                   if plan_details['plan_number'] != template_plan]
    
    # Run the plans (either in parallel or sequentially)
    if run_parallel:
        print(f"\nRunning {len(plan_numbers)} plans in parallel...")
        results = RasCmdr.compute_parallel(
            plan_number=plan_numbers,
            max_workers=max_workers,
            num_cores=num_cores,
            clear_geompre=True
        )
    else:
        print(f"\nRunning {len(plan_numbers)} plans sequentially...")
        results = {}
        for plan_number in plan_numbers:
            print(f"  Running plan {plan_number}...")
            result = RasCmdr.compute_plan(plan_number, clear_geompre=True)
            results[plan_number] = result
    
    print("\nExecution results:")
    for plan, success in results.items():
        print(f"  Plan {plan}: {'Successful' if success else 'Failed'}")
    
    # Use template geometry for mesh cell identification
    geom_number = template_geom
    geom_hdf_path = ras.geom_df.loc[ras.geom_df['geom_number'] == geom_number, 'hdf_path'].values[0]
    
    # Find the nearest mesh cell to the point of interest
    mesh_cells_gdf = HdfMesh.get_mesh_cell_points(geom_hdf_path)
    
    # Calculate distances from the point to all mesh cells
    distances = mesh_cells_gdf.geometry.apply(lambda geom: geom.distance(point_of_interest))
    
    # Find the index of the minimum distance
    nearest_idx = distances.idxmin()
    
    # Get the cell ID for results extraction
    mesh_cell_for_results = mesh_cells_gdf.loc[nearest_idx, 'cell_id']
    mesh_name = mesh_cells_gdf.loc[nearest_idx, 'mesh_name']
    
    print(f"\nNearest cell ID: {mesh_cell_for_results}")
    print(f"Distance: {distances[nearest_idx]:.2f} units")
    print(f"Mesh area: {mesh_name}")
    
    # Extract and store results
    all_results = {}
    max_ws_values = []
    
    # Get results for each plan
    for plan_details in sensitivity_plans:
        plan_number = plan_details['plan_number']
        adjustment = plan_details['adjustment']
        shortid = plan_details['shortid']
        
        # Get the results for this plan
        try:
            results_xr = HdfResultsMesh.get_mesh_cells_timeseries(plan_number)
            
            # Extract water surface data for the specific cell
            ws_data = results_xr[mesh_name]['Water Surface'].sel(cell_id=int(mesh_cell_for_results))
            
            # Convert to DataFrame for easier handling
            ws_df = pd.DataFrame({
                'time': ws_data.time.values,
                'water_surface': ws_data.values
            })
            
            # Store in results dictionary
            all_results[plan_number] = {
                'adjustment': adjustment,
                'shortid': shortid,
                'df': ws_df,
                'max_water_surface': ws_df['water_surface'].max()
            }
            
            # Store the maximum water surface value for summary
            max_ws = ws_df['water_surface'].max()
            max_ws_values.append({
                'plan_number': plan_number,
                'shortid': shortid,
                'adjustment': adjustment,
                'max_water_surface': max_ws
            })
            
            print(f"Plan {plan_number} ({shortid}, adjustment {adjustment:+.3f}): Max Water Surface = {max_ws:.2f}")
            
            # Save individual time series to CSV
            plan_csv_path = results_dir / f"timeseries_{shortid}.csv"
            ws_df.to_csv(plan_csv_path, index=False)
            
        except Exception as e:
            print(f"Error extracting results for plan {plan_number}: {str(e)}")
    
    # Create a summary DataFrame for maximum water surface values
    max_ws_df = pd.DataFrame(max_ws_values)
    max_ws_df = max_ws_df.sort_values('adjustment')
    
    # Create a friendly land use description for plot titles
    landuse_desc = ", ".join(landuse_types)
    if len(landuse_desc) > 40:  # Truncate if too long
        landuse_desc = landuse_desc[:37] + "..."
    
    # Create a friendly region description
    if table_num is not None:
        region_desc = f"Table {table_num} Only"
    else:
        region_desc = "All Region Tables"
    
    # File name prefix
    file_prefix = f"manning_sensitivity_region_landuse_{table_prefix}{landuse_shortid}"
    
    # Save the summary to CSV
    csv_path = results_dir / f"{file_prefix}_summary.csv"
    max_ws_df.to_csv(csv_path, index=False)
    print(f"\nSummary saved to: {csv_path}")
    
    # Plot the time series results
    plt.figure(figsize=(14, 8))
    
    # Color map for different adjustments (blue for negative, red for positive)
    cmap = plt.cm.coolwarm
    adjustment_range = max(abs(max_ws_df['adjustment'].min()), abs(max_ws_df['adjustment'].max()))
    norm = plt.Normalize(-adjustment_range, adjustment_range)
    
    # Plot time series for each plan
    for _, row in max_ws_df.iterrows():
        plan_number = row['plan_number']
        adjustment = row['adjustment']
        shortid = row['shortid']
        
        # Get the result data
        result_data = all_results[plan_number]
        df = result_data['df']
        
        color = cmap(norm(adjustment))
        label = f"{shortid} ({adjustment:+.3f})"
        
        plt.plot(df['time'], df['water_surface'], label=label, color=color, 
                 linewidth=2 if adjustment == 0 else 1)
    
    # Add plot details
    plt.title(f'Water Surface Sensitivity to Region Manning\'s Roughness ({region_desc})\nfor Land Uses: {landuse_desc}\nCell ID: {mesh_cell_for_results}')
    plt.xlabel('Time')
    plt.ylabel('Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Save the plot
    plot_path = results_dir / f"{file_prefix}_timeseries.png"
    plt.tight_layout()
    plt.savefig(plot_path)
    print(f"Time series plot saved to: {plot_path}")
    
    # Create a summary plot of max water surface vs adjustment
    plt.figure(figsize=(10, 6))
    plt.scatter(max_ws_df['adjustment'], max_ws_df['max_water_surface'], c=max_ws_df['adjustment'], 
                cmap='coolwarm', norm=norm, s=100)
    
    # Add a trend line
    z = np.polyfit(max_ws_df['adjustment'], max_ws_df['max_water_surface'], 1)
    p = np.poly1d(z)
    plt.plot(max_ws_df['adjustment'], p(max_ws_df['adjustment']), "r--", alpha=0.8)
    
    # Calculate the slope for annotation
    slope = z[0]
    
    # Add plot details
    plt.title(f'Maximum Water Surface vs Region Manning\'s Roughness Adjustment ({region_desc})\nfor Land Uses: {landuse_desc}')
    plt.xlabel('Manning\'s Roughness Adjustment')
    plt.ylabel('Maximum Water Surface Elevation (ft)')
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add annotation with sensitivity information
    plt.annotate(f'Sensitivity: {slope:.4f} ft per unit n\n(or {slope*0.01:.4f} ft per 0.01 change in n)',
                xy=(0.05, 0.05), xycoords='axes fraction',
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
    
    # Save the summary plot
    summary_plot_path = results_dir / f"{file_prefix}_summary.png"
    plt.tight_layout()
    plt.savefig(summary_plot_path)
    print(f"Summary plot saved to: {summary_plot_path}")
    
    # Show the plots
    plt.show()
    
    return {
        'all_results': all_results,
        'max_ws_summary': max_ws_df,
        'mesh_cell_id': mesh_cell_for_results,
        'mesh_name': mesh_name,
        'landuse_types': landuse_types,
        'landuse_codes': landuse_codes,
        'region_tables': region_tables_to_adjust
    }
```
